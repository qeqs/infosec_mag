```python
import os
import io
import math
import sys
import imagehash as imagehash

from Crypto import Random
from Crypto.Cipher import AES
from Crypto.Protocol.KDF import PBKDF2
from PIL import Image, ImageDraw
from IPython.display import display
```

```python
def get_key_PBKDF2(key):
        salt = os.urandom(8)  # 64-bit
        return PBKDF2(key, salt, 32)  # 256-bit key

def pad(s):
    return s + b"\0" * (AES.block_size - len(s) % AES.block_size)

def encrypt(message, key):
    message = pad(message)
    iv = Random.new().read(AES.block_size)
    cipher = AES.new(key, AES.MODE_CBC, iv)
    return iv + cipher.encrypt(message)

def decrypt(ciphertext, key):
    iv = ciphertext[:AES.block_size]
    cipher = AES.new(key, AES.MODE_CBC, iv)
    plaintext = cipher.decrypt(ciphertext[AES.block_size:])
    return plaintext.rstrip(b"\0")
```

```python
def get_hamming_distance(decripted_data, deformed_image_hash):
    res = []
    for i in range(0, len(decripted_data)):
        temp = str(deformed_image_hash[i]).encode()
        var = sum(1 for t in zip(temp, decripted_data[i]) if t[0] != t[1])
        res.append(var)
    return res

def binarize(data):
    v = []
    bytes = [ord(chr(b)) for b in data]

    for b in bytes:
        for i in range(7, -1, -1):
            v.append((b >> i) & 0x1)

    return v

def set_bit(n, i, x):
    mask = 1 << i
    n &= ~mask
    if x:
        n |= mask
    return n

def steganography_proccess(fragment_image, _hash):
    width, height = fragment_image.size
    conv = fragment_image.convert("RGBA").getdata()
    max_size = width * height * 3.0 / 8 / 1024

    v = binarize(_hash)
    while len(v) % 3:
        v.append(0)

    payload_size = len(v) / 8 / 1024.0
    if payload_size > max_size:
        raise "Cannot embed. File too large"

    steg_img = Image.new('RGBA', (width, height))
    data_img = steg_img.getdata()

    idx = 0
    for h in range(height):
        for w in range(width):
            (r, g, b, a) = conv.getpixel((w, h))
            if idx < len(v):
                r = set_bit(r, 0, v[idx])
                g = set_bit(g, 0, v[idx + 1])
                b = set_bit(b, 0, v[idx + 2])
            data_img.putpixel((w, h), (r, g, b, a))
            idx = idx + 3

    return steg_img
```

```python
key = 'password'
cryptographic_key = get_key_PBKDF2(key)

original_image = Image.open("temp.png")


array_images = []
n = 32
width, height = original_image.size

fix_width = n * math.ceil(width / n)
fix_height = n * math.ceil(height / n)

fix_image = Image.new('RGBA', (fix_width, fix_height))
fix_image.paste(original_image, (0, 0))

result_image = Image.new('RGBA', (fix_width, fix_height))

x = 0
y = 0
for h in range(0, fix_height + 1, n):
    for w in range(0, fix_width + 1, n):
        fragment_image = fix_image.crop((w, h, w + n, h + n))
        
        _hash = imagehash.average_hash(fragment_image)
        encrypted_hash = encrypt(str(_hash).encode(), cryptographic_key)
        fragment_image = steganography_proccess(fragment_image, encrypted_hash)
        
        array_images.append(fragment_image)
        result_image.paste(fragment_image, (x, y))
        if (x + n) <= fix_width:
            x = x + n
        else:
            x = 0
            y = y + n


result_image = result_image.crop((0, 0, width, height))
# result_image.show()
display(result_image)
result_image.save("lab_1_image.png", "PNG")

```

```python
# array of bits into a binary file
def assemble(v):
    mass_bytes = ""
    payload_size = 48

    length = len(v)
    for idx in range(0, int(len(v) / 8)):
        byte = 0
        for i in range(0, 8):
            if (idx * 8 + i < length):
                byte = (byte << 1) + v[idx * 8 + i]
        mass_bytes = mass_bytes + chr(byte)

    mass_bytes = [ord(b) for b in mass_bytes]
    mass_bytes = mass_bytes[: payload_size]

    output_data = b''
    for b in mass_bytes:
        output_data = b''.join([output_data, bytes([b])])
    return output_data

def extract(steg_image):
    width, height = steg_image.size
    conv = steg_image.convert("RGBA").getdata()

    v = []
    for h in range(height):
        for w in range(width):
            (r, g, b, a) = conv.getpixel((w, h))
            v.append(r & 1)
            v.append(g & 1)
            v.append(b & 1)

    data_out = assemble(v)
    return data_out
```

```python

deformed_image = Image.open("temp_deform.png")
deformed_images_hash = []
array_deformed_images = []

fix_image = Image.new('RGBA', (fix_width, fix_height))
fix_image.paste(deformed_image, (0, 0))

for h in range(0, fix_height + 1, n):
    for w in range(0, fix_width + 1, n):
        fragment_image = fix_image.crop((w, h, w + n, h + n))
        array_deformed_images.append(fragment_image)
        deformed_images_hash.append(imagehash.average_hash(fragment_image))

decripted_data = [decrypt(extract(sample), cryptographic_key) for sample in array_images]


hamming_distance = get_hamming_distance(decripted_data, deformed_images_hash)
threshold_value = len(decripted_data[0]) * 0.25

# draw_differences(array_deformed_images, hamming_distance, threshold_value)

print("Hamming distance : %s" % hamming_distance)
```
